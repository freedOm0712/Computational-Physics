```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
```

```python
#number of sample of points
N=800

#samle spacing
T=1.0/N

t=np.linspace(0,N*T, N)
y=2 * np.sin(2 * np.pi * t) + 2 * np.cos(2 * np.pi * t) + 2 * np.sin(2 * np.pi * t)
```

```python
#perform DFT using scipy library
yf=fft(y)
xf=np.linspace(0, 1/(2*T), N//2)

#perform DFT using numpy library
ynp=np.fft.fft(y)

titles = ['x','scipy','numpy']
data=[titles]+list(zip(xf,yf,ynp))

#print result
print ('Result from sciy and numpy')
for i, d in enumerate(data):
    line='|'.join (str(x).ljust(12) for x in d)
    print(line)
```

```python
#Plot the result (Original function)
plt.figure(figsize=(10,6))
plt.title('$f(x)=2 \sin (2\pi t)+2 \cos(2\pi t)+2 \sin (2\pi t)$')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.plot(t,y,label='original function')
plt.legend()
plt.grid()
```

```python
#Plot the result (FFT from scipy library)
plt.figure(figsize=(10,6))
plt.title('FFT')
plt.xlabel('Frequency (Hz)')
plt.ylabel ('Amplitude')
plt.plot(t,yf,color='k',label='scipy fft')
plt.plot(t,ynp,color='r',linestyle=':',label = ' numpy fft')
plt.legend()
plt.grid()
```

```python
#Plot the result (Inverse Fourier Transform)
plt.figure(figsize=(10,6))
plt.title('Inverse FFT')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(xf,2/N*np.abs(yf[0:N//2]),color='k',label='scipy fft')
plt.plot(xf, 2/N*np.abs(ynp[0:N//2]),color='r',linestyle=':',label='numpy fft')
plt.legend()
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def y_function(x):
    return 2 * np.sin(2 * np.pi * x) + 2 * np.cos(2 * np.pi * x) + 2 * np.sin(2 * np.pi * x)

# Generate x values
x = np.linspace(0, 2, 500)  # From 0 to 2 (2 cycles for better visualization)

# Calculate y values
y = y_function(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$y = 2\sin(2\pi x) + 2\cos(2\pi x) + 2\sin(2\pi x)$', color='blue')
plt.title('Plot of $y = 2\sin(2\pi x) + 2\cos(2\pi x) + 2\sin(2\pi x)$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid()
plt.legend()
plt.show()
```
