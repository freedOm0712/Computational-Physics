**QUESTION 1a**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import sympy as sp
from scipy.integrate import simps
```

```python
def f(x):
    return np.exp(-x) * np.cos(x)
```

```python
#Simpson's rule
xnew=np.linspace(0,1,1000)
ynew=f(xnew)
area_simps=simps(ynew, x=xnew)
print (f'The area calculated using Simpson\'s rule is {area_simps:0.6f}')

#Monte Carlo 
width = 1-0
samples = np.random.uniform(low=0,high=width,size=10000)
area_MC=f(samples).mean()*width
error=np.std(samples*width)/np.sqrt(samples.size)
print (f'The area calculated using Monte Carlo method is {area_MC:0.6f} ± {error:0.6f}')

#Trapezoid rule
area_trapezoid = np.trapz(ynew, xnew)
print (f'The area calculated using Trapeziod rule is {area_trapezoid:0.6f}')

#analytical values
x = sp.symbols('x') #define the symbol x in the function
func = sp.exp(-x) * sp.cos(x)
area_analytical = sp.integrate(func, (x, 0, 1)) # Perform the symbolic integration
area_analytical_numeric = area_analytical.evalf() # Convert the result to a numerical value
print(f'The analytical value of the area is {area_analytical_numeric:0.6f}')
```

```python
#plot the result
plt.figure(figsize=(10,6))
plt.fill_between(xnew,0,ynew, alpha =0.1)
plt.title('Question 1a : Area calculated using Simpsom\'s method, Monte Carlo, Trapezoid rule and analytical method')
plt.text(0.25,0.85,f'Area from Simps is {area_simps:0.6f}')
plt.text(0.25,0.65,f'Area from Monte Carlo Intergration is {area_MC:0.6f} $\pm$ {error:0.6f}')
plt.text(0.25,0.45,f'Area from Trapezoid rule is {area_trapezoid:0.6f}')
plt.text(0.25,0.25,f'Area from analytical method is {area_analytical_numeric:0.6f}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(xnew, ynew)
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
from scipy.integrate import simps
import matplotlib.pyplot as plt
```

```python
def fnormal(x):
    return (np.exp(-x))/x
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.integrate import simps

def fnormal(x):
    return np.exp(-x) / x

# Monte Carlo Integration 
x_samp = norm.rvs(size=50000000)  
x_samp = x_samp[(x_samp > 1e-6) & (x_samp < 1)]  
p_of_x = norm.pdf(x_samp)  
vals = fnormal(x_samp) / p_of_x  
area_mc = vals.mean()  
error_mc = np.std(vals) / np.sqrt(vals.size)  

# Simpson's Rule Integration 
xs = np.linspace(1e-6, 1, 1000000)  
ys = fnormal(xs)
area_simps = simps(ys, x=xs)  

# Plot the results
plt.figure(figsize=(10,6))
plt.plot(xs, ys, label='Function', lw=3)
plt.fill_between(xs, 0, ys, alpha=0.1)
plt.text(0.85, 0.85, f'MC Area = {area_mc:0.2f} ± {error_mc:0.2f}')
plt.text(0.25, 0.25, f'Simps Area = {area_simps:0.2f}')
plt.legend()
plt.grid(True)
plt.show()
```
